```python
import queue
class BinaryTreeNode:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
def Print(root,k,d=0):
    if root is None:
        return 
    if d==k:
        print(root.data)
        return
    Print(root.left,k,d+1)
    Print(root.right,k,d+1)
    return
def PrintNodes(root,Node,k):
    if root is None:
        return -1
    if root.data==Node:
        Print(root,k)
        return 0
    x=PrintNodes(root.left,Node,k)
    if x==k:
        print(root.data)
    y=PrintNodes(root.right,Node,k)
    if y==k:
        print(root.data)
        
    
def LevelInput(list):
    index=0
    length = len(list)
    if length<=0 or list[0]==-1:
        return None
    q=queue.Queue()
    root=BinaryTreeNode(list[index])
    index+=1
    q.put(root)
    while not q.empty():
        curr=q.get()
        left=list[index]
        if left!=-1:
            curr.left=BinaryTreeNode(left)
            q.put(curr.left)
        index+=1
        right=list[index]
        if right!=-1:
            curr.right=BinaryTreeNode(right)
            q.put(curr.right)
        index+=1
    
    return root  
        
li=[int(x) for x in input().split()]
Node=int(input())
k=int(input())
root=LevelInput(li)
PrintNodes(root,Node,k)
```
